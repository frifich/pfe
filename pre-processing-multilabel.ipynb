```python
import os
import gc
import numpy as np 
```

```python
# data files 
data = list(os.walk("./data"))[0][2]
exec_traces = [open(os.path.join("./data", f"{f}")) for f in data]
```

```python
for idx in range(len(exec_traces)):
    
    # reset solutions and objectives 
    solutions = list()
    objectives = list()
    single_solutions = list()
    
    # building the solutions and objectives lists 
    for line in exec_traces[idx]:
    
        # solutions
        solution = str()
        try:
            for ref in line.split(","): 
                ref = ref.split("(")
                op = ref[0]
                loc = ref[1].split(";")[:2]
                for i in range(len(loc)):
                    try :
                        if loc[i][:5] == "Class" :
                            del loc[i]
                            i -= 1
                    except:
                        continue
                    if loc[i] == "": 
                        del loc[i]
                solution += " ".join([op] + loc) + " "
            solutions.append(solution) 
        except: 
            continue
        # objectives 
        try:
            next_line = [-1 * float(obj) for obj in next(exec_traces[idx]).split("(")[1].split(")")[0].split(",")][:6]
            assert len(next_line) == 6
        except:
            solutions.pop()
            continue
        objectives.append(next_line)
    
    # close data file 
    exec_traces[idx].close()
    
    # must have same number of solutions and objectives 
    try: 
        assert len(solutions) == len(objectives)
    except AssertionError: 
        continue
    
    print(f"File {exec_traces[idx].name} has {len(solutions)} solutions and {len(objectives)} objectives\n")
    
    # define threshold 
    objectives = np.array(objectives)
    medians = [np.median(objectives[:, i]) for i in range(6)]
    threshold = [max(medians[i], 0) for i in range(6)]
    
    # no threshold
#     threshold = [0] * 6
    
    #building the single solutions with the multilabel
    count = 0
    for i, obj in enumerate(objectives):
        label = [1 if obj[x] > threshold[x] else 0 for x in range(6)]
        if sum(label) == 0:
            continue
        else: 
            single_solutions.append(solutions[i] + ", ".join(map(str, label)))
            
    print(f"total of {len(single_solutions)} single solutions have improved at least one objective\n")
    
    # removing redundant solutions
    single_solutions = list(set(single_solutions))
    print(f"For a total of {len(single_solutions)} single solutions\n")
    
    # save the multilabeled solutions 
    with open(os.path.join("./processed_data/", f"FinalSolutions-multilabel-opencsv.txt"), 'a') as f:
        f.write("\n".join(single_solutions) + "\n")
    
    # forcing garbbage collector 
    del objectives
    del solutions
    del single_solutions
    gc.collect()
```
